### Univariate Analysis

```python
df = data[data['price'] < 1500]
sns.boxplot(data=df,x='price')
import matplotlib.pyplot as plt 
plt.show()
```

```python
#price distribution
data['price']
```

```python
### price distribution 
plt.figure(figsize=(6,3))
sns.histplot(data=df,x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
import matplotlib.pyplot as plt
plt.show()
```

### Feature Engineering 

```python
# data [price per beds]
df['price per beds']=df['price']/df['beds']
```

### Bi Variable Analysis one variable dependency in another variable

```python
df.columns
```

```python
### price dependency for neighbourhood
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
import matplotlib.pyplot as plt
plt.show()
```

```python
### number of review and reletionship
plt.title('Locality and Review Dependency')
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.figure(figsize=(8,5))

import matplotlib.pyplot as plt
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df,vars=['price', 'minimum_nights', 'number_of_reviews','reviews_per_month','availability_365'],hue='room_type')
import matplotlib.pyplot as plt
plt.show()
```

```python
### Geograhical Distribution Airbnb listing
plt.title('Geograhical Distribution Airbnb listing')
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.figure(figsize=(10,7))
import matplotlib.pyplot as plt
plt.show()
```

```python
# Heat Map - correlation of one variable with other numerical columns
corr =df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','beds']].corr()
corr
```

```python
sns.heatmap(data=corr,annot=True)
plt.figure(figsize=(8,6))
import matplotlib.pyplot as plt
plt.show()
```
